```python
T = int(input())
for i in range(T):
    N = int(input())
    meetings = []
    for i in range(N):
        S, E = map(int, input().split())
        meetings.append((S, E))
    meetings.sort(key=lambda x: x[1])
    count = 0
    last_end = float('-inf')
    for start, end in meetings:
        if start >= last_end:
            count += 1
            last_end = end
    print(count)
```
